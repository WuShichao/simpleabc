```python
from ABC import abc
```

```python
class MyModel(object):
    
    def __init__(self):
        pass
    
    def draw_theta(self):
        mu = stats.uniform.rvs(-2,4)
        sigma = stats.uniform.rvs(0.01,2)
        
        return mu,sigma
    
    def generate_data(self, theta):
        return stats.norm.rvs(loc=theta[0],scale=theta[1],size=100)
    
    def summary_stats(self, data):
        per25 = mean(data)
        per75 = std(data)
        
        return [per25,per75] 
    
    def distance_function(self, data, synth_data):

        return sqrt((data[0]-synth_data[0])**2 + (data[1]-synth_data[1])**2)
```

```python
mu_real, sig_real = 0.5, 1.4
model = MyModel()
data = stats.norm.rvs(loc=mu_real, scale=sig_real, size=100)
```

```python
post, reject, accept, tot = abc(model, data, min_particles=500)
```

```python
post
```

```python
re_mu = [x[0] for x in reject]
re_sigma = [x[1] for x in reject]
ac_mu = [x[0] for x in post]
ac_sigma = [x[1] for x in post]
```

```python
ax=subplot(111)
plot(re_mu, re_sigma, 'o', c='0.5',mec='0.5')
axvline(mu_real, ls='--',color='orange')
axhline(sig_real, ls='--',color='orange')
plot(ac_mu,ac_sigma,'o', c='k')

```
