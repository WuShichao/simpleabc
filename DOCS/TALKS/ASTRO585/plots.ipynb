```python
from astropy.io import ascii
```

```python
sim = ascii.read('benchmark.dat')
norm = ascii.read('benchmark_norm.dat')
simC = ascii.read('benchmark_C.dat')
normC = ascii.read('benchmark_norm_C.dat')
```

```python
print sim
print norm
print simC
print normC
```

```python
plot(sim['col1'],sim['col2'],'o-', label='Simple Planet Model')
plot(norm['col1'],norm['col2'],'o-', label='Normal Distribtuion')
xticks(range(0,9),['Serial']+['%s' % x for x in range(1,9)])
title('Cyberstar RCC Cluster; Single Node; 8 Cores 4GB Shared Memory', fontsize=20)
xlabel('Number of Spawned Processes')
ylabel('Time [s]')
grid()
legend()
savefig('time.eps')
```

```python
plot(sim['col1'][1:],sim['col3'][0]/sim['col3'][1:],'o-', label='Simple Planet Model')
plot(norm['col1'][1:],norm['col3'][0]/norm['col3'][1:],'o-', label='Normal Distribtuion')
title('Cyberstar RCC Cluster; Single Node; 8 Cores 4GB Shared Memory', fontsize=20)
xlabel('Number of Spawned Processes')
ylabel('Normalized Speed-Up')
grid()
legend()
savefig('speedup.eps')
```

```python
plot(simC['col1'],simC['col2'],'o-', label='Simple Planet Model')
plot(normC['col1'],normC['col2'],'o-', label='Normal Distribtuion')
xticks(range(0,9),['Serial']+['%s' % x for x in range(1,9)])
title('Domino Data Labs (Amazon Cloud); Single Node; 4 Cores 4GB Shared Memory', fontsize=20)
xlabel('Number of Spawned Processes')
ylabel('Time [s]')
grid()
legend()
savefig('time-C.eps')
figure()
plot(simC['col1'][1:],simC['col3'][0]/simC['col3'][1:],'o-', label='Simple Planet Model')
plot(normC['col1'][1:],normC['col3'][0]/normC['col3'][1:],'o-', label='Normal Distribtuion')
title('Domino Data Labs (Amazon Cloud); Single Node; 4 Cores 4GB Shared Memory', fontsize=20)
xlabel('Number of Spawned Processes')
ylabel('Normalized Speed-Up')
grid()
legend()
savefig('speedup-C.eps')
```

```python
plot(sim['col1'][1:],sim['col3'][0]/sim['col3'][1:],'bo-', label='Simple Planet Model RCC')
plot(simC['col1'][1:],simC['col3'][0]/simC['col3'][1:],'bs--', label='Simple Planet Model Cloud')
plot(norm['col1'][1:],norm['col3'][0]/norm['col3'][1:],'go-', label='Normal Distribtuion RCC')
plot(normC['col1'][1:],normC['col3'][0]/normC['col3'][1:],'gs--', label='Normal Distribtuion Cloud')
xlim(1,4)
legend()
xlabel('Number of Spawned Processes')
ylabel('Normalized Speed-Up')
grid()
savefig('speedup-vs.eps')
```

```python
plot(sim['col1'],sim['col2'],'bo-', label='Simple Planet Model RCC')
plot(simC['col1'],simC['col2'],'bs--', label='Simple Planet Model Cloud')
plot(norm['col1'],norm['col2'],'go-', label='Normal Distribtuion RCC')
plot(normC['col1'],normC['col2'],'gs--', label='Normal Distribtuion Cloud')
legend()
xlim(0,4)
xlabel('Number of Spawned Processes')
ylabel('Time [s]')
grid()
savefig('time-vs.eps')
```
